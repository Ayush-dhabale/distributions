```python
import numpy as np
import seaborn as sns
from scipy.stats import f
```

### Normal Distribution

```python
normal_data = np.random.normal(loc= 3, scale= 1, size= 10000)
```

here, loc --> population mean

scale ---> std deviation

```python
sns.histplot(normal_data, kde= True, color= "red")
```

### Standard Normal Distribution

This data distribution have mean equals to 0 and std deviation equals to 1

```python
std_normal_data = np.random.standard_normal(size=10000)
```

```python
sns.histplot(std_normal_data, kde= True, color= "black")
```

### Log Normal Distribution

```python
log_normal_data = np.random.lognormal(mean=3,sigma=1,size= 1000)
```

```python
sns.histplot(log_normal_data,kde= True, color='purple')
```

log of log normal data is  a normal distribution

```python
sns.histplot(np.log(log_normal_data),kde=True, color='red')
```

### Power Law Distribution

```python
power_data = np.random.power(a=4,size=10000)
```

```python
sns.histplot(power_data,kde=True, color='black')
```

#### Pareto Distribution

```python
pareto_data = np.random.pareto(a=4,size=1000)
```

```python
sns.histplot(pareto_data,kde=True, color='blue')
```

#### Exponential Distribution

```python
exponential_data = np.random.exponential(scale=4,size=1000)
```

```python
sns.histplot(exponential_data, kde=True, color='green')
```

### Binomial Distribution

```python
binomial_data = np.random.binomial(10,0.2,10000)
```

```python
sns.histplot(binomial_data,kde=True,color='pink')
```

### Poission Distribution

```python
poission_data = np.random.poisson(lam=24, size=10000)
```

```python
sns.histplot(poission_data,kde=True, color='red')
```
